```python
class TiendaOnline:
    def __init__(self,inventario,clientes,carritos):
        
        self.inventario = inventario
        self.clientes = clientes
        self.ventas_totales = 0
        self.carritos = carritos
  
    def agregar_producto(self,nombre,precio,cantidad):
        self.nombre = ((nombre).strip()).lower()
        self.precio = float(precio)
        self.cantidad = int(cantidad)

        #Creo un diccionario con los parametros pasados en el metodo
        diccionario_agregar = {'nombre': nombre,'precio': precio,'cantidad': cantidad}
        
        # Recorro la lista de diccionarios,le asigno un indice 
        # para acceder a cada diccionario y busco por el nombre si el nombre del producto existe.
    
        for posicion,valor in enumerate(self.inventario):
            #Si existe, actualizo la cantidad
            if valor.get('nombre')== nombre:
                self.inventario[2] = cantidad
                break
            else:
                print('No existe,se puede agregar')
                self.inventario.append(diccionario_agregar)
                break
        print(self.inventario)


    def ver_inventario(self):
        for posicion,producto in enumerate(self.inventario):
            nombre = (producto.get('nombre')).title()
            precio = producto.get('precio')
            cantidad = producto.get('cantidad')
            print(f'Nombre: {nombre}  Precio: $ {precio}  Cantidad: {cantidad}')
         
    def buscar_producto(self,nombre):
        nombre = ((nombre).strip()).lower()
        self.nombre = nombre
        for posicion,dicc in enumerate(self.inventario):
            #Si existe, lo muestro
            if dicc.get('nombre') == nombre:
                print('Nombre:',dicc.get('nombre'),'Precio: $',dicc.get('precio'))
                return dicc.get('precio')
        print('No se ha encontrado el producto')
        

       

    
    def actualizar_stock(self,nombre,cantidad):

        self.nombre = nombre
        self.cantidad = cantidad

        for posicion,dicc in enumerate(self.inventario):
            #Si existe, actualizo la cantidad
            if dicc.get('nombre')== nombre:
                dicc['cantidad'] = dicc.get('cantidad') + cantidad
                nombre = (dicc.get('nombre')).title()
                precio = dicc.get('precio')
                cantidad = dicc.get('cantidad')
                #print(f'Nombre: {nombre}  Precio: ${precio}  Cantidad: {cantidad}' )
                return
        print('El producto no se encuentra en el inventario')  
        



    def eliminar_producto(self, nombre):
        self.nombre = nombre
        #Recorro la lista de diccionarios y si encuentro el producto,elimino por el indice de la lista
        for posicion,dicc in enumerate(self.inventario):
            if dicc.get('nombre')== nombre:
                self.inventario.pop(posicion)
                break
            else:
                print(f'El producto {nombre} no está en el inventario')

    def calcular_valor_inventario(self):
        #Recorro cada diccionario de la lista inventario y dentro de cada uno calculo el valor de ese 
        #diccionario y al final sumo los valores de todos los diccionarios
        cantidad=0
        precio =0
        total_dicc =0
        for posicion,dicc in enumerate(self.inventario):
            cantidad = dicc.get('cantidad')
            precio = dicc.get('precio')
            total_dicc = total_dicc + (cantidad * precio)
        
        print( 'El valor total del inventario es: $',total_dicc)
            
    def realizar_compra(self,carrito={},lista_carritos_compra=[],nombre_cliente=""):
        #Recorro el inventario y guardo en un diccionario los productos y los precios
        #Muestro un input con el  dicc_productos_precio
        # le pido que escriba el nombre del producto
        #muestro otro input para que me diga la cantidad
        #le pregunto si quiere seguir comprando,en cuyo caso vuelvo a mostrar los inputs anteriores.
        #Cuando ya decide no comprar mas,le pregunto finalmente si es cliente.
        # En caso afirmativo,lo compruebo y si es cliente guardo en el carrito su compra y su nombre
        #creo un diccionario con los productos, el precio y el total      
        
        dicc_productos_precio ={}
        for posicion,dicc in enumerate(self.inventario):

            dicc_productos_precio.setdefault(dicc.get('nombre'),dicc.get('precio'))
        
        producto_elegido = ((input(f'Elija un producto de la lista y escriba su nombre: {dicc_productos_precio}' )).strip()).lower()
        cantidad_elegida = int(input(f'Ha elegido comprar {producto_elegido},cuantas unidades quiere?.Indique las unidades que desea en números')) 
        
        if producto_elegido!='' and cantidad_elegida!='':
            #llamo al metodo y hago una busquedan por nombre y guardo el precio
            precio = self.buscar_producto(producto_elegido)
            
            #Agrego al carrito de compra el producto,el precio y la cantidad
            dicc_carrito= {'nombre': producto_elegido,'precio': precio,'cantidad': cantidad_elegida}
            
            #Modifico el inventario con las cantidades
            self.actualizar_stock(producto_elegido,-cantidad_elegida)
            
            #agrego este diccionario a una lista de diccionarios por si quiere seguir comprando
            lista_carritos_compra.append(dicc_carrito)
            
            
        continuar_compra =(input('Desea seguir agregando productos a su cesta?. Pulse Y para continuar')).lower()
        if continuar_compra == 'y': #El cliente elige continuar con la tecla Y
            #Volvemos a llamar al metodo compra pero esta vez le pasamos el carrito que ya tiene productos y la lista
            self.realizar_compra(dicc_carrito,lista_carritos_compra)
            
        #####calculo el importe totall del carrito y lo añado al carrito
        cantidad=0
        precio =0
        total_carrito =0
        ############## le pregunto si es cliente, si la respuesta es si lo guardo con su nombre
        resp=((input('Es usted cliente?,Si lo es pulse la letra Y ,de lo contrario pulse la letra N')).strip()).lower()
        if resp=='y':
            nombre_cliente=input('Escriba su nombre y su primer apellido')
            if self.buscar_cliente(nombre_cliente)==True:
                nombre_cliente = nombre_cliente
                print('el cliente esta en la base de datos ',nombre_cliente)
            else:
                print('Usted no es cliente')
                nombre_cliente == ''
        #######Agrego el carrito a la lista de carritos
        for posicion,carrito in enumerate(lista_carritos_compra):
            #print('recorriendo la lista de carritos',carrito)
            cantidad = carrito.get('cantidad')
            precio = carrito.get('precio')
            total_carrito = round(total_carrito + (cantidad * precio),2)
            carrito.setdefault('total',total_carrito)

        #Si el cliente existe,registramos su compra
        if resp =='y':
            self.registrar_compra(nombre_cliente,carrito)
            
        print( 'El valor total del carrito es: $',total_carrito)
      
        
        
        
    def procesar_pago(self,cantidad_total=0,cantidad_aportada=0):
        # validamos lo que mete el usuario, si hay algo, si es un numero, si es mayor que la compora,etc
        # Para ello creo una funcion dentro del metodo que chequea los inputs
        
        cantidad_total = input('Indique la cantidad total de la compra: ')
        cantidad_aportada = input('Ingreso que aporta : ')    

        if self.validar_entrada_datos(cantidad_total,cantidad_aportada)==True:
            cambio= round(float(cantidad_aportada),2) - round(float(cantidad_total),2)
            print('Su cambio es de $',cambio)

    ##Metodo creado para validar 2 importes    
    def validar_entrada_datos(self,num1,num2):
        import string
        self.num1 = num1
        self.num2 = num2
        if num1==''or num2=='':
            print('Debe introducir una cantidad')
            return False
        elif num1.isdigit()== False or num2.isdigit()== False:
            print('Debe introducir un valor numerico')
            return False

        elif int(num1) < 0 or int(num2) < 0:
            print('Los importes deben ser positivos')
            return False
        elif int(num2) < int(num1):
            print('La cantidad a pagar debe ser igual ó superior al total de la compra')
            return False
        else:
            return True
        
    def agregar_cliente(self, nombre, email):
        self.nombre = (nombre).lower()
        self.email = email
        dicc_cliente={}

        if nombre == "" or email =="" :
            print('Debe introducir un nombre y un email válido')
        else:
            dicc_cliente = {'nombre': nombre,'email': email}
            self.clientes.append(dicc_cliente)

    def ver_clientes(self):

        #El self.clientes al igual que inventario será una lista de diccionarios de clientes
        #Recorremos la lista y printeamos cada cliente que tenemos
        for posicion,cliente in enumerate(self.clientes):
            nombre = cliente.get('nombre') 
            email = cliente.get('email')
            print(f'Nombre: {nombre}    Email: {email}')

    def buscar_cliente(self,nombre):
        nombre = (nombre).lower()
        self.nombre = nombre
        for posicion,cliente in enumerate(self.clientes):
            #Si existe,me salgo y devuelvo true
            if cliente.get('nombre') == nombre:
                return True
            
        print('El cliente no está en nuestra base de datos,no está registrado')
        
    def registrar_compra(self, nombre_cliente, carrito):
        self.nombre_cliente = nombre_cliente
        self.carrito = carrito
        #Busco el cliente y si existe agrego la compra a su carrito
        #Si no existe,devuelvo print indicandolo y no hago nada
        if self.buscar_cliente(nombre_cliente)==True:
            #Agrego al carrito
            print('el cliente existe')
            carrito.setdefault('cliente',nombre_cliente)
            self.carritos.append(carrito)
        
    def ver_compras_cliente(self, nombre_cliente):
        self.nombre_cliente = nombre_cliente

        #Busco si el cliente lo tenemos guardado y si es asi empiezo a 
        #recorrer la lista de carritos para buscar su nombre y mostrar
        #sólo las compras que ese cliente haya efectuado

        if self.buscar_cliente(nombre_cliente)==True:
           
            #busco en la lista de carritos compra y muestro los que el campo cliente coincida con el nombre
            #si coincide lo agrego a una lista historial_compra_cliente
            #Procedo a calcular el total de la compra del cliente
            historial_compra_cliente=[]
            #Printeo el encabezado de la lista con el nombre del cliente
            print(f'Historial de compras del cliente {(nombre_cliente).title()}')
            for posicion,carrito in enumerate(self.carritos):
                if carrito.get('cliente')==nombre_cliente:
                    historial_compra_cliente.append(carrito)    
                    
            #Recorro la lista y la muestro mas 'bonita'
            total_compra_cliente =0
            for pos,compra in enumerate(historial_compra_cliente):  
                total_compra_cliente = total_compra_cliente + compra['total']
                print('Nombre Producto:',compra['nombre'],'Precio: $',compra['precio'],'Cantidad Comprada:',compra['cantidad'],'Total: $',compra['total'])
                      
            total_compra_cliente = round(total_compra_cliente,2)

            print(f'Este cliente ha comprado en nuestra tienda por un total de ${total_compra_cliente}')
    
    def calcular_ventas_totales(self):
        ventas_totales = 0
        for pos,carrito in enumerate(self.carritos):
            ventas_totales = ventas_totales + carrito['total']
            
        ventas_totales = round(ventas_totales,2)
        print(f'Las ventas de la tienda ascienden a un total de: ${ventas_totales}.')
        
                    
    def buscar_producto_regex(self, patron):
        import re
        self.patron = patron
        productos_encontrados = []
        producto =''
        for pos,producto in enumerate(self.inventario):
            for i in producto:
                coincidencia =re.search(patron,producto['nombre'])
                if coincidencia != None:
                    #print(producto)
                    productos_encontrados.append(producto)
                    break
               
        #Muestro la lista con los resultados
        for pos,producto in enumerate(productos_encontrados):  
            print('Nombre Producto:',producto['nombre'],'Precio: $',producto['precio'],'Cantidad:',producto['cantidad'])
                      
                

            
```
